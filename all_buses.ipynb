```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as Ec
from  selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
import time
```

```python
import pandas as pd 
a=pd.read_csv(r"D:\project\andra_routes.csv")
a
```

```python
#--------------------------------------------------------Andra details------------------------------------------------------------

driver=webdriver.Chrome()
wait=WebDriverWait(driver,20)
driver.maximize_window()

ap_bus_name=[]
ap_bus_type=[]
ap_bus_departing_time=[]
ap_bus_duration=[]
ap_bus_reaching_time=[]
star_rating=[]
price_list=[]
seat_availability=[]
route_name=[]
route_link=[]
for i,k in a.iterrows():
    links=k["andra_keys"]
    routs=k["andra_routes"]
    driver.get(links)
    time.sleep(5)
    websites=driver.find_elements(By.XPATH,f"//a[contains(@href,'{links}')]")
    time.sleep(2)
    for website in websites:
        website.click()
        time.sleep(2)
        WebDriverWait(driver,20).until(Ec.presence_of_all_elements_located((By.XPATH,"//*[@class='result-sec']")))
    try:
        inside=(driver.find_element(By.XPATH,"//div[@class='button']"))
        inside.click()
    except:
       print("button not available")
    time.sleep(3)
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(10, document.body.scrollHeight);")
        time.sleep(8) 
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        #time.sleep(2)
    bus_name=driver.find_elements(By.XPATH,"//div[@class='travels lh-24 f-bold d-color']")    
    bus_type=driver.find_elements(By.XPATH,"//div[@class='bus-type f-12 m-top-16 l-color evBus']")    
    starting=driver.find_elements(By.XPATH,"//*[@class='dp-time f-19 d-color f-bold']")    
    duration=driver.find_elements(By.XPATH,"//*[@class='dur l-color lh-24']")    
    ending=driver.find_elements(By.XPATH,"//*[@class='bp-time f-19 d-color disp-Inline']")    
    try:
        rating=driver.find_elements(By.XPATH,"//div[@class='clearfix row-one']//div[@class='column-six p-right-10 w-10 fl']")    
    except:
        continue
    price=driver.find_elements(By.XPATH,"//*[@class='fare d-block']")    
    seats_avail=driver.find_elements(By.XPATH,"//div[contains(@class,'seat-left')]")     
    for bus in bus_name:
        ap_bus_name.append(bus.text)
        route_name.append(routs)
        route_link.append(links)
    for bus_types in bus_type:
        ap_bus_type.append(bus_types.text)
    for starting_times in starting:
        ap_bus_departing_time.append(starting_times.text)
    for travelling_time in duration:
        ap_bus_duration.append(travelling_time.text)
    for ending_time in ending:
        ap_bus_reaching_time.append(ending_time.text)
    for ratings in rating:
        star_rating.append(ratings.text)
    for pricing in price:
        price_list.append(pricing.text)
    for seats in seats_avail:
        seat_availability.append(seats.text)
print("success")
```

```python
data1={
"bus_name":ap_bus_name,
"bus_type":ap_bus_type,
"starting_time":ap_bus_departing_time,
"time_duration":ap_bus_duration,
"reaching_time":ap_bus_reaching_time,
"rating":star_rating,
"fare_price":price_list,
"seats":seat_availability,
"route_name":route_name,
"route_link":route_link}
andra=pd.DataFrame(data1)
location=r"D:\total project\andra_routes.csv"
andra.to_csv(location,index=False)
```
